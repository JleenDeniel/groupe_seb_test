```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
```

```python
values = np.array(range(100))
```

```python
for i in range(100):
    values[i] = random.randint(0,100)
```

```python
normalized = values / np.linalg.norm(values)
```

```python
plt.plot(np.array(range(100)), normalized)
```

```python
def count_mean(vector):
    sum = 0
    for num in vector:
        sum += num
    return sum/len(vector)
```

```python
def count_variance(vector):
    mean = count_mean(vector)
    sum = 0
    for num in vector:
        sum += (num - mean) ** 2
    return sum/len(vector)
```

```python
def count_std(vector):
    return count_variance(vector) ** (1/2)
```

```python
print("Математическое ожидание:" + str(count_mean(normalized)))
print("Дисперсия:" + str(count_variance(normalized)))
print("Стандартное отклонение:" + str(count_std(normalized)))
```
